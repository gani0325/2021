## 1. 직선 그리기

```python
# 다양한 직선 그리기
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/가은/2022/img/flower1.jpg')

cv2.line(img, (20, 50), (50, 50), (255,0,0))   # 파란색 1픽셀 선
cv2.line(img, (60, 50), (100, 50), (0,255,0),  20, cv2.LINE_4)  # 초록색 4 연결 선
cv2.line(img, (200, 50), (250, 50), (0,0,255), 10)  # 빨간색 10픽셀 선

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 2. 사각형 그리기

```python
# 사각형 그리기

import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/가은/2022/img/flower1.jpg')

# 좌상, 우하 좌표로 사각형 그리기, 선 두께는 default 1
cv2.rectangle(img, (50, 50), (120, 150), (255,0,0) )        
# 우하, 좌상 좌표로 사각형 그리기, 선 두께 10
cv2.rectangle(img, (100, 80), (120, 100), (0,255,0), 10 )  
# 우상, 좌하 좌표로 사각형 채워 그리기 ---①
cv2.rectangle(img, (140, 100), (200, 150), (0,0,255), -1 )  

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 3. 다각형 그리기

```python
# 다각형 그리기(draw_poly)

import numpy as np                          # 좌표 표현을 위한 numpy 모듈  ---①
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/가은/2022/img/house.jpg')

# Numpy array로 좌표 생성 ---②
# 번개 모양 선 좌표
pts1 = np.array([[50,50], [150,150], [100,140],[200,240]], dtype=np.int32) 
# 삼각형 좌표
pts2 = np.array([[250,50], [150,200], [270,100]], dtype=np.int32) 
# 삼각형 좌표
pts3 = np.array([[100,250], [0,200], [200,100]], dtype=np.int32) 
# 5각형 좌표
pts4 = np.array([[150,50], [250,150], [200,250], [100,250], [50,150]],\
                 dtype=np.int32) 

# 다각형 그리기 ---③
cv2.polylines(img, [pts1], False, (255,0,0))       # 번개 모양 선 그리기
cv2.polylines(img, [pts2], False, (0,0,0), 10)     # 3각형 열린 선 그리기 ---④
cv2.polylines(img, [pts3], True, (0,0,255), 10)    # 3각형 닫힌 도형 그리기 ---⑤
cv2.polylines(img, [pts4], True, (0,0,0), cv2.LINE_4)          # 5각형 닫힌 도형 그리기

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 4. 원, 타원, 호 그리기

```python
# 원, 타원, 호 그리기(draw_circle.py)

import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/가은/2022/img/house.jpg')

# 원점(150,150), 반지름 100 ---①
cv2.circle(img, (150, 150), 100, (255,0,0))     

# 원점(400,150), 반지름 50, 채우기 ---③
cv2.circle(img, (400, 150), 50, (0,0,255), -1)  

# 원점(50,300), 반지름(50), 회전 0, 0도 부터 360도 그리기 ---④
cv2.ellipse(img, (50, 300), (50, 50), 0, 0, 360, (0,0,255))    

# 원점(325, 300), 반지름(75,50) 납작한 타원 그리기 ---⑦
cv2.ellipse(img, (325, 300), (75, 50), 0, 0, 360, (0,255,0))    
# 원점(450,300), 반지름(50,75) 홀쭉한 타원 그리기 ---⑧
cv2.ellipse(img, (450, 300), (50, 75), 0, 0, 360, (255,0,255))    

# 원점(350,425), 홀쭉한 타원 45도 회전 후 아랫 반원 그리기 ---⑪
cv2.ellipse(img, (150, 265), (50, 75), 45, 0, 180, (0,0,255))    
# 원점(400,425), 홀쭉한 타원 45도 회전 후 윗 반원 그리기 ---⑫
cv2.ellipse(img, (200, 225), (50, 75), 45, 181, 360, (255,0,0))   


cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 5. 글쓰기

```python
# 글 쓰기

import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/가은/2022/img/house.jpg')

# sans-serif small
cv2.putText(img, "Alpaca", (50, 30), cv2.FONT_HERSHEY_PLAIN, 1, (0, 0,0))                
# sans-serif normall X2  ---①
cv2.putText(img, "cute girl!", (10, 110), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,0,250)) 

# serif small
cv2.putText(img, "Image Processing", (50, 180), cv2.FONT_HERSHEY_COMPLEX_SMALL, \
            1, (0, 0,0))            

# hand-wringing sans-serif
cv2.putText(img, "I'm pretty", (50, 230), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, \
            1, (0, 0,0)) 

# sarif + italic
cv2.putText(img, "very hungry", (100, 170), \
            cv2.FONT_HERSHEY_COMPLEX | cv2.FONT_ITALIC, 1, (0, 0,0)) 

cv2_imshow(img)
cv2.waitKey()
cv2.destroyAllWindows()
```
